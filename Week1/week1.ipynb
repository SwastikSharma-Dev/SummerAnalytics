## GRADED Questions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
```

```python
df = pd.read_csv("cars.csv")
```

### 1. What is name of car that has the highest horsepower?  

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
max_hp_car = df.loc[df['horsepower'].idxmax()]
print("Car with highest horsepower:", max_hp_car['name'])
```

### 2. How many cars have mpg ≥ 35?

```python
high_mpg_count = (df['mpg'] >= 35).sum()
print("Number of cars with mpg >= 35:", high_mpg_count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = filtered['origin'].mode()[0]
print("Most common origin:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
japan_acc_mean = df[df['origin'] == 'japan']['acceleration'].mean()
print("Mean acceleration of Japanese cars:", round(japan_acc_mean, 2))
```

### 5. Which year had the highest average mpg?

```python
best_year = df.groupby('model_year')['mpg'].mean().idxmax()
print("Year with highest average mpg:", best_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
df['hp_per_weight'] = df['horsepower'] / df['weight']
median_mpg = df['mpg'].median()
filtered = df[df['mpg'] > median_mpg]
best_ratio = filtered['hp_per_weight'].max()
best_cars = filtered[filtered['hp_per_weight'] == best_ratio]
print("Cars with best hp/weight and above-median mpg:")
print(best_cars[['name', 'mpg', 'hp_per_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns

avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
index_str = df.index.astype(str)
df['base_name'] = index_str.to_series().str.extract(r'^([^\d]+)').iloc[:, 0].str.strip()

car_counts = df['base_name'].value_counts()
repeated_names = car_counts[car_counts > 1].index
repeated_df = df[df['base_name'].isin(repeated_names)]

consistency = repeated_df.groupby('base_name').agg(
    mpg_std=('mpg', 'std'),
    count=('mpg', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

consistent_models = consistency[(consistency['mpg_std'] < 1.0) & (consistency['count'] >= 2)]
consistent_models = consistent_models.sort_values(by=['count', 'avg_mpg'], ascending=[False, False])

print("Consistent car models (low mpg std across years):")
print(consistent_models)
```
